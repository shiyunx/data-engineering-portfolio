```python
# Import libraries
import numpy as np
import pandas as pd
from google.colab import files

# Create dictionary (keys = column names)
data = {
    "StudentID": [100100, 100102, 100103, 100104],
    "StudentName": ["Alpha", "Beta", "Gamma", "Delta"],
    "Physics": [65, 70, 32, 60],
    "Geography": [54, 62, 71, np.nan], # Missing
    "Math": [78, 80, 96, -10], # Invalid
    "Semester": ["2025S1", "2025S1", "2025S1", "2025-S1"] # Inconsistent
}

# Convert dictionary into a pandas dataframe table
df_raw = pd.DataFrame(data)

# Read dataframe
display(df_raw)

# Save the dataframe as a csv
csv_filename = "raw_marks.csv"
df_raw.to_csv(csv_filename, index=False)

# Download the csv
files.download(csv_filename)
```
